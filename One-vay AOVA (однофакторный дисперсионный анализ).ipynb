```python
import pandas as pd
from scipy.stats import f_oneway
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://stepik.org/media/attachments/lesson/8083/genetherapy.csv')
```

```python
f_oneway(*[list(frame) for (group, frame) in df.groupby('Therapy')['expr']])
```

# Вывод
p - значение меньше 0.05, поэтому отвергаем нулевую гипотезу и принимаем альтернативную: различия в группах не случайно и они из разных генеральных совокупностей

```python
im = sns.boxplot( x=df["Therapy"], y=df["expr"] )
im.set(xlabel='Группы',
 ylabel='Уровень эксперессии',
 title='Выражееность генов по группам')
plt.show()
```
